# <span style="color:pink"> Introducción a Experimentos con MLflow</span>
En este cuaderno realizaremos un ejercicio sencillo orientado a comprender cómo gestionar experimentos de machine learning utilizando MLflow. El objetivo principal es que puedas observar, de manera práctica, cómo registrar métricas, parámetros y resultados de diferentes ejecuciones de un modelo, facilitando así el monitoreo y la comparación entre experimentos.

En proyectos reales de Ciencia de Datos, es común probar múltiples configuraciones de modelos hiperparámetros, algoritmos, datasets, etc. Sin una herramienta adecuada, este proceso se vuelve difícil de rastrear y reproducir. Aquí es donde MLflow aporta un valor clave.

## ¿Reemplaza MLflow a GridSearch o RandomSearch?

Es importante aclarar que **MLflow no reemplaza** técnicas como **Grid Search** o **Random Search**.

Estas técnicas, disponibles por ejemplo en scikit-learn, se usan para **buscar automáticamente los mejores hiperparámetros**.

En cambio, **MLflow se encarga de**:

- Registrar cada experimento  
- Guardar parámetros y métricas  
- Comparar resultados entre ejecuciones  
- Facilitar la reproducibilidad  

**Relación entre ellos:**

- **GridSearch / RandomSearch** → optimizan modelos  
- **MLflow** → organiza y monitorea experimentos  

De hecho, pueden usarse juntos: puedes ejecutar un GridSearch y registrar cada intento en MLflow.

### Ventajas de usar MLflow

- Trazabilidad completa de experimentos  
- Comparación visual de resultados  
- Reproducibilidad, saber exactamente cómo se obtuvo un modelo  
- Registro centralizado de modelos y métricas  
- Facilita el trabajo en equipo  

### ¿Qué es MLflow?

**MLflow es una plataforma open-source para gestionar el ciclo de vida del machine learning**. Sus componentes principales son:

- **Tracking** → registro de experimentos, lo que usaremos en este cuaderno  
- **Projects** → empaquetado de código reproducible  
- **Models** → gestión y despliegue de modelos  
- **Registry** → versionado de modelos  

En este ejercicio nos enfocaremos en **MLflow Tracking**, que permite registrar automáticamente:

- Parámetros del modelo  
- Métricas como accuracy  
- Artefactos como archivos, modelos y gráficos

# Ejecución de experimento con MLflow (servidor remoto)

En esta celda ejecutaremos nuestro primer experimento de machine learning utilizando MLflow, registrando los resultados en un servidor remoto.

Antes de iniciar el experimento, configuramos la conexión con el servidor de MLflow. Esto permitirá que todos los parámetros, métricas y modelos se almacenen de forma centralizada y puedan ser visualizados posteriormente en la interfaz web.

````python
# Definir experimento
mlflow.set_experiment("experimento con logistic regression")

# Iniciar experimento
with mlflow.start_run(run_name="ejemplo-logistic-regression") as run:

    # Cargar datos
    X, y = load_iris(return_X_y=True)

    # Definir parámetros
    params = {
        "C": 0.1,
        "random_state": 42
    }

    # Registrar parámetros
    mlflow.log_params(params)

    # Entrenar modelo
    lr = LogisticRegression(**params).fit(X, y)

    # Predicciones y evaluación
    y_pred = lr.predict(X)
    accuracy = accuracy_score(y, y_pred)

    # Registrar métrica
    mlflow.log_metric("accuracy", accuracy)

    # Guardar modelo — usar 'name' en vez de 'artifact_path' (deprecado en MLflow 3.x)
    # y formato 'skops' para evitar el warning de pickle
    mlflow.sklearn.log_model(
        lr,
        name="model",
        serialization_format=mlflow.sklearn.SERIALIZATION_FORMAT_SKOPS
    )

    print(f"✅ Run ID: {run.info.run_id}")
    print(f"📊 Accuracy: {accuracy:.4f}")
    print(f"📦 Artifact URI: {mlflow.get_artifact_uri()}")
````

## Si lo desea guardar en formato Joblib

````python
import joblib

# Entrenar modelo
lr = LogisticRegression(**params).fit(X, y)

# Guardar con joblib
joblib.dump(lr, "model.joblib")

# Registrar métricas y el archivo
mlflow.log_metric("accuracy", accuracy)
mlflow.log_artifact("model.joblib", artifact_path="model")

print(f"📦 Artifact URI: {mlflow.get_artifact_uri()}")
````

````python
mlflow.end_run()
````

# Búsqueda Exhaustiva de Hiperparámetros con GridSearchCV

En este experimento aprenderás a usar **GridSearchCV** para encontrar la mejor 
combinación de hiperparámetros de un modelo de clasificación, registrando cada 
intento en MLflow para su análisis posterior.

## ¿Qué es GridSearchCV?
GridSearchCV prueba **todas las combinaciones posibles** de los valores que tú 
defines para cada hiperparámetro. Es como una cuadrícula (grid) donde cada celda 
es una combinación diferente.

Por ejemplo, si defines:
- C = [0.001, 0.01, 0.1, 1, 10, 100]     → 6 valores
- max_iter = [100, 200, 500]              → 3 valores  
- solver = ["lbfgs", "saga"]             → 2 valores

GridSearch probará 6 × 3 × 2 = **36 combinaciones**, entrenando el modelo 
con validación cruzada de 5 folds en cada una → 36 × 5 = **180 entrenamientos**.

## ¿Qué registraremos en MLflow?
- Los hiperparámetros del mejor modelo encontrado (best_C, best_solver, etc.)
- El accuracy promedio de validación cruzada de cada combinación (como curva por step)
- El accuracy del mejor modelo sobre el conjunto de entrenamiento
- El modelo final serializado en formato skops

## ¿Cuándo usar GridSearch?
✅ Cuando el espacio de búsqueda es pequeño (pocas combinaciones)  
✅ Cuando necesitas garantía de haber probado todas las opciones  
❌ No es ideal cuando tienes muchos hiperparámetros o rangos amplios — 
   el número de combinaciones crece exponencialmente.

````python
import mlflow
import os
import urllib3
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
from sklearn.metrics import accuracy_score
from scipy.stats import uniform

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)
os.environ['MLFLOW_TRACKING_INSECURE_TLS'] = 'true'
mlflow.set_tracking_uri("https://unab-n8n.duckdns.org/mlflow")

# Cargar datos (una sola vez)
X, y = load_iris(return_X_y=True)

# ── Definir experimento ──────────────────────────────────────────
mlflow.set_experiment("busqueda-de-hiperparametros")

# ════════════════════════════════════════════════════════════════
# RUN 1: GridSearchCV — búsqueda exhaustiva
# ════════════════════════════════════════════════════════════════
with mlflow.start_run(run_name="GridSearchCV") as run:

    # Tags del run (estas sí existen)
    mlflow.set_tags({
        "modelo":        "LogisticRegression",
        "dataset":       "Iris",
        "tipo_busqueda": "GridSearchCV",
        "autor":         "Adrian Diaz",
        "version":       "1.0",
        "framework":     "scikit-learn",
    })
    
    
    param_grid = {
        "C":            [0.001, 0.01, 0.1, 1, 10, 100],
        "max_iter":     [100, 200, 500],
        "solver":       ["lbfgs", "saga"],
    }

    gs = GridSearchCV(
        LogisticRegression(random_state=42),
        param_grid,
        cv=5,
        scoring="accuracy",
        n_jobs=-1,
        verbose=0,
    )
    gs.fit(X, y)

    # Registrar parámetros de búsqueda
    mlflow.log_param("metodo",          "GridSearchCV")
    mlflow.log_param("cv_folds",        5)
    mlflow.log_param("total_combinaciones", len(gs.cv_results_["params"]))

    # Registrar los mejores parámetros encontrados
    mlflow.log_params({f"best_{k}": v for k, v in gs.best_params_.items()})

    # Métricas del mejor modelo
    mlflow.log_metric("best_cv_accuracy",  gs.best_score_)
    mlflow.log_metric("best_cv_std",       gs.cv_results_["std_test_score"][gs.best_index_])

    # Accuracy sobre el conjunto completo (entrenamiento)
    y_pred = gs.best_estimator_.predict(X)
    mlflow.log_metric("train_accuracy", accuracy_score(y, y_pred))

    # Registrar todas las combinaciones como métricas con step
    for step, (params, mean, std) in enumerate(zip(
        gs.cv_results_["params"],
        gs.cv_results_["mean_test_score"],
        gs.cv_results_["std_test_score"],
    )):
        mlflow.log_metric("cv_mean_accuracy", mean, step=step)
        mlflow.log_metric("cv_std_accuracy",  std,  step=step)

    # Guardar mejor modelo
    mlflow.sklearn.log_model(
        gs.best_estimator_,
        name="best_model",
        serialization_format=mlflow.sklearn.SERIALIZATION_FORMAT_SKOPS,
    )

    print(f"✅ GridSearch — Mejor accuracy CV: {gs.best_score_:.4f}")
    print(f"   Mejores parámetros: {gs.best_params_}")
    print(f"   Run ID: {run.info.run_id}")


````

# Búsqueda Aleatoria de Hiperparámetros con RandomizedSearchCV

En este experimento usarás **RandomizedSearchCV** como alternativa más eficiente 
a GridSearch, especialmente útil cuando el espacio de hiperparámetros es grande 
o cuando algunos parámetros son continuos (no discretos).

## ¿Qué es RandomizedSearchCV?
En lugar de probar todas las combinaciones posibles, RandomizedSearchCV toma 
**n muestras aleatorias** del espacio de búsqueda. Esto tiene dos ventajas clave:

1. **Puede usar distribuciones continuas** — por ejemplo, en vez de probar 
   C = [0.1, 1, 10], puedes muestrear C de una distribución uniforme entre 
   0.001 y 100, cubriendo valores que Grid nunca probaría.

2. **Controlas el presupuesto computacional** — con `n_iter=30` siempre 
   entrenarás exactamente 30 combinaciones, sin importar qué tan grande sea 
   el espacio de búsqueda.

## ¿Qué registraremos en MLflow?
- El número de iteraciones realizadas (n_iter)
- Los mejores hiperparámetros encontrados
- La curva de accuracy ordenada de menor a mayor (para ver la mejora progresiva)
- El modelo final en formato skops

## Comparación con GridSearchCV
| Aspecto              | GridSearchCV        | RandomizedSearchCV     |
|----------------------|---------------------|------------------------|
| Combinaciones        | Todas               | n_iter aleatorias      |
| Parámetros continuos | No                  | Sí (distribuciones)    |
| Garantía del óptimo  | Sí (dentro del grid)| No, pero aproximado    |
| Costo computacional  | Alto                | Controlable            |

## ¿Cuándo usar RandomSearch?
✅ Cuando el espacio de búsqueda es grande  
✅ Cuando tienes parámetros continuos (como la tasa de aprendizaje)  
✅ Cuando el tiempo de entrenamiento es limitado  
❌ No garantiza haber probado la mejor combinación exacta

````python
# ════════════════════════════════════════════════════════════════
# RUN 2: RandomizedSearchCV — búsqueda aleatoria
# ════════════════════════════════════════════════════════════════
with mlflow.start_run(run_name="RandomizedSearchCV") as run:

    param_dist = {
        "C":        uniform(loc=0.001, scale=100),   # distribución continua
        "max_iter": [100, 200, 300, 500],
        "solver":   ["lbfgs", "saga", "newton-cg"],
    }

    rs = RandomizedSearchCV(
        LogisticRegression(random_state=42),
        param_dist,
        n_iter=30,          # probar 30 combinaciones aleatorias
        cv=5,
        scoring="accuracy",
        random_state=42,
        n_jobs=-1,
        verbose=0,
    )
    rs.fit(X, y)

    # Registrar parámetros de búsqueda
    mlflow.log_param("metodo",          "RandomizedSearchCV")
    mlflow.log_param("cv_folds",        5)
    mlflow.log_param("n_iter",          30)

    # Registrar los mejores parámetros encontrados
    mlflow.log_params({f"best_{k}": v for k, v in rs.best_params_.items()})

    # Métricas del mejor modelo
    mlflow.log_metric("best_cv_accuracy", rs.best_score_)
    mlflow.log_metric("best_cv_std",      rs.cv_results_["std_test_score"][rs.best_index_])

    # Accuracy sobre el conjunto completo
    y_pred = rs.best_estimator_.predict(X)
    mlflow.log_metric("train_accuracy", accuracy_score(y, y_pred))

    # Registrar todas las iteraciones como métricas con step
    # Ordenadas por mean_test_score para ver la curva de mejora
    sorted_idx = np.argsort(rs.cv_results_["mean_test_score"])
    for step, idx in enumerate(sorted_idx):
        mlflow.log_metric("cv_mean_accuracy", rs.cv_results_["mean_test_score"][idx], step=step)
        mlflow.log_metric("cv_std_accuracy",  rs.cv_results_["std_test_score"][idx],  step=step)

    # Guardar mejor modelo
    mlflow.sklearn.log_model(
        rs.best_estimator_,
        name="best_model",
        serialization_format=mlflow.sklearn.SERIALIZATION_FORMAT_SKOPS,
    )

    print(f"✅ RandomSearch — Mejor accuracy CV: {rs.best_score_:.4f}")
    print(f"   Mejores parámetros: {rs.best_params_}")
    print(f"   Run ID: {run.info.run_id}")
````

````python
mlflow.end_run()
````
